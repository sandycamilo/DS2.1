### Import and explore the dataset. Recall the load_breast_cancer() method will return an object that contains the data in .data, the labels in .target, and the column names in .feature_names attributes.

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd 
```

```python
BreastCancer = load_breast_cancer()
print(BreastCancer.data)
```

```python
print(BreastCancer.data.shape)
```

```python
breastCancer = pd.DataFrame(BreastCancer.data)
breastCancer.columns = BreastCancer.feature_names
breastCancer['T'] = BreastCancer.target

breastCancer.head()
```

```python
# columns-feature_names & data
# target - T ~ labels
breastCancer.describe()
```

### Build a Correlation Heatmap using Seaborn to check for each feature's correlation with the labels.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 

#calculate correlation matrix
corr = breastCancer.corr()

# plot heatmap
sns.heatmap(corr,
    xticklabels=corr.columns,
    yticklabels=corr.columns
)
```

### Scale and transform the data using a StandardScaler() object and any appropriate methods it contains.

```python
from sklearn.preprocessing import StandardScaler

#define scaler
scaler = StandardScaler()
#transform data
scaled = scaler.fit_transform(breastCancer)
scaled
```

### Split the newly scaled data into training and testing sets using train_test_split().

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression


X = scaled
# print(X)
y = breastCancer['T']
# print(y)

# split X and y into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

### Create an SVC() object, which can be found in sklearn.svm

```python
from sklearn.svm import SVC
from sklearn import svm

#define suport vector classifier 
clf = svm.SVC(kernel='linear', C = 1.0)
```

### Fit the model to the scaled data.

```python
clf.fit(X,y)
```

### Use your validation data to check the accuracy metrics for your model.

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
y_test.to_numpy() 
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)

#TN = 53
#FP = 0
#FN = 0 
#TP = 90
```

```python
# from sklearn.metrics import accuracy_score
# accuracy_score(y_test, y_pred)

#accuracy score
TN = 53
FP = 0
FN = 0 
TP = 90

Accuracy = ((TP + TN) / float(TP + TN + FP + FN))

print(Accuracy)
```

```python
#precision score

Precision = TP / float(TP + FP)

print(Precision)
```

```python
#Recall 

Recall = TP / float(FN + TP)
print(Recall)
```

```python
#f1 score
F1_score = 2*Precision*Recall/float(Precision+Recall)

print(F1_score)
```
